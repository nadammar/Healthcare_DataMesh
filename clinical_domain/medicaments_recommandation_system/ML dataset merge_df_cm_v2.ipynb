## **Giving an ID to each medication**

```python
mc=medication["MedicationText"].unique()
```

```python
IDm=[]
for j in range(len(medication.MedicationText)) :
  for i in range(len(mc)) :
    if medication["MedicationText"][j] == mc[i]:
      IDm.append(i+1)
```

```python
medication["ID"]=IDm
```

```python
medication
```

```python
medication=medication.loc[:,["ID","MedicationText","PatientUID","Date"]]
```

```python
medication 
```

## **Giving an ID to each encounter**

```python
me=condition["ConditionText"].unique()
```

```python
IDe=[]
for j in range(len(condition.ConditionText)) :
  for i in range(len(me)) :
    if condition["ConditionText"][j] == me[i]:
      IDe.append(i+1)
```

```python
condition["ID"]=IDe
```

```python
condition=condition.loc[:,["ID","ConditionText","PatientUID","Date"]]
```

```python
merge_df=condition.merge(medication,on=["Date","PatientUID"],how='left',suffixes=('_Condition','_Medication'))
merge_df
```

##**Cleaning Data**

```python
merge_df.duplicated().value_counts()
```

```python
merge_df=merge_df.drop_duplicates()
```
